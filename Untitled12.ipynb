```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  # Check if 'year_1' column exists in _df_15, if not replace 'Birth Year' with actual column name containing year data
  xs = series['Birth Year'] if 'Birth Year' in series else series['year_1']
  ys = series['acha_12months_any_18NUM']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_15.sort_values('Birth Year', ascending=True) # Replace 'Birth Year' if necessary

for i, (series_name, series) in enumerate(df_sorted.groupby('phq9_2')):
  _plot_series(series, series_name, i)
  fig.legend(title='phq9_2', bbox_to_anchor=(1, 1), loc='upper left')

sns.despine(fig=fig, ax=ax)
plt.xlabel('Birth Year') # Replace 'Birth Year' with actual column name if modified in _plot_series
_ = plt.ylabel('acha_12months_any_18NUM')
```
